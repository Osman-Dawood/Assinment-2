Task 1 - Tensor Manipulations: Practice tensor operations and manipulations in Numpy.


```python
import numpy as np

# Create two sample tensors
tensor_a = np.array([[1, 2, 3], [4, 5, 6]])
tensor_b = np.array([[7, 8, 9], [10, 11, 12]])
print("Tensor A:\n", tensor_a)
print("Tensor B:\n", tensor_b)

# Element-wise addition of tensors
tensor_add = tensor_a + tensor_b
print("\nElement-wise Addition:\n", tensor_add)

# Element-wise multiplication of tensors
tensor_mul = tensor_a * tensor_b
print("\nElement-wise Multiplication:\n", tensor_mul)

# Reshape a tensor (2x3 to 3x2)
reshaped_tensor = tensor_a.reshape(3, 2)
print("\nReshaped Tensor (2x3 to 3x2):\n", reshaped_tensor)

# Slicing and indexing tensors
sliced_tensor = tensor_a[0, :]  # Extract the first row
indexed_element = tensor_a[1, 2]  # Access element in row 2, column 3
print("\nSliced Tensor (First Row):\n", sliced_tensor)
print("Indexed Element (Row 2, Column 3):", indexed_element)

# Combining tensors (concatenate along axis 0)
combined_tensor = np.concatenate((tensor_a, tensor_b), axis=0)
print("\nCombined Tensor (Concatenated along Axis 0):\n", combined_tensor)

# Splitting tensors (split tensor_a into two parts along axis 1)
split_tensors = np.split(tensor_a, 3, axis=1)
print("\nSplit Tensors (Split along Axis 1):\n", split_tensors)

# Basic mathematical operations on tensors
sum_tensor = np.sum(tensor_a)  # Sum of all elements
mean_tensor = np.mean(tensor_a)  # Mean of all elements
sqrt_tensor = np.sqrt(tensor_a)  # Square root of each element

print("\nSum of Elements in Tensor A:", sum_tensor)
print("Mean of Elements in Tensor A:", mean_tensor)
print("Square Root of Tensor A:\n", sqrt_tensor)
```

Task 2 -Neural Network Foundations: Construct fundamental blocks of neural networks using Keras.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Step 1: Define a simple neural network model
model = Sequential([
    Dense(32, activation='relu', input_shape=(10,)),  # Input layer
    Dense(16, activation='relu'),                     # Hidden layer
    Dense(1, activation='sigmoid')                    # Output layer
])

# Step 2: Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 3: Print the model summary
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

# Save the model architecture visualization as an image
plot_model(model, to_file="model_architecture.png", show_shapes=True, show_layer_names=True)

# This will save an image named "model_architecture.png" in your working directory.
```

Task 3 - Classification Tasks: Build and train a fully connected neural network for classification tasks using first MNIST and then IMDb datasets by repeating the eight steps below:

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist, imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding, GlobalAveragePooling1D # Import GlobalAveragePooling1D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences


# --- Load and Preprocess MNIST ---
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()
x_train_mnist, x_test_mnist = x_train_mnist / 255.0, x_test_mnist / 255.0  # Normalize
y_train_mnist, y_test_mnist = to_categorical(y_train_mnist), to_categorical(y_test_mnist)  # One-hot encode

# --- Define MNIST Model ---
mnist_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# --- Compile MNIST Model ---
mnist_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# --- Train MNIST Model ---
mnist_history = mnist_model.fit(x_train_mnist, y_train_mnist, validation_split=0.2, epochs=10, batch_size=32)

# --- Evaluate MNIST Model ---
mnist_loss, mnist_accuracy = mnist_model.evaluate(x_test_mnist, y_test_mnist)
print("MNIST Test Loss:", mnist_loss)
print("MNIST Test Accuracy:", mnist_accuracy)

# --- Plot MNIST Performance ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(mnist_history.history['loss'], label='Training Loss')
plt.plot(mnist_history.history['val_loss'], label='Validation Loss')
plt.title('MNIST Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(mnist_history.history['accuracy'], label='Training Accuracy')
plt.plot(mnist_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('MNIST Accuracy')
plt.legend()
plt.show()

# --- Load and Preprocess IMDb ---
(x_train_imdb, y_train_imdb), (x_test_imdb, y_test_imdb) = imdb.load_data(num_words=10000)
x_train_imdb = pad_sequences(x_train_imdb, maxlen=200)
x_test_imdb = pad_sequences(x_test_imdb, maxlen=200)

# --- Define IMDb Model ---
imdb_model = Sequential([
    Embedding(10000, 64, input_length=200),
    GlobalAveragePooling1D(),  # Add GlobalAveragePooling1D layer
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
# --- Compile IMDb Model ---
imdb_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# --- Train IMDb Model ---
imdb_history = imdb_model.fit(x_train_imdb, y_train_imdb, validation_split=0.2, epochs=5, batch_size=32)

# --- Evaluate IMDb Model ---
imdb_loss, imdb_accuracy = imdb_model.evaluate(x_test_imdb, y_test_imdb)
print("IMDb Test Loss:", imdb_loss)
print("IMDb Test Accuracy:", imdb_accuracy)

# --- Plot IMDb Performance ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(imdb_history.history['loss'], label='Training Loss')
plt.plot(imdb_history.history['val_loss'], label='Validation Loss')
plt.title('IMDb Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(imdb_history.history['accuracy'], label='Training Accuracy')
plt.plot(imdb_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('IMDb Accuracy')
plt.legend()
plt.show()
```
